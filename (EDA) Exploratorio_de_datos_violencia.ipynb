```python
import pandas as pd

violencia = pd.read_csv(r"/content/violencia_de_genero.csv", sep=",", decimal=",", encoding="utf-8")
```

```python
violencia.head(2)
```

```python
violencia.info()
```

```python
violencia.drop(['pac_hos_','con_fin_','sust_vict','hora_hecho','escenario','zona_conf','nmun_resi','nom_eve'],axis = 'columns', inplace=True)
```

```python
violencia.head(2)
```

```python
violencia['Cantidad casos'] = ''
violencia.loc[violencia['Orden']>0,'Cantidad casos'] = 1
```

```python
violencia.head(2)
```

```python
violencia['Grup edad clasif'] = ''
violencia.loc[violencia['Grupo edad'] == '0 a 6','Grup edad clasif'] = 1
violencia.loc[violencia['Grupo edad'] == '7  a 11','Grup edad clasif'] = 2
violencia.loc[violencia['Grupo edad'] == '12 a 17','Grup edad clasif'] = 3
violencia.loc[violencia['Grupo edad'] == '18 a 28','Grup edad clasif'] = 4
violencia.loc[violencia['Grupo edad'] == '29 a 59','Grup edad clasif'] = 5
violencia.loc[violencia['Grupo edad'] == '60 y mas','Grup edad clasif'] = 6

violencia['Ciclo de vida clasif'] = ''
violencia.loc[violencia['Ciclo de vida'] == 'Adolescencia','Ciclo de vida clasif'] = 1
violencia.loc[violencia['Ciclo de vida'] == 'Adultez','Ciclo de vida clasif'] = 2
violencia.loc[violencia['Ciclo de vida'] == 'Infancia','Ciclo de vida clasif'] = 3
violencia.loc[violencia['Ciclo de vida'] == 'Jovenes','Ciclo de vida clasif'] = 4
violencia.loc[violencia['Ciclo de vida'] == 'Persona Mayor','Ciclo de vida clasif'] = 5
violencia.loc[violencia['Ciclo de vida'] == 'Primera infancia','Ciclo de vida clasif'] = 6

violencia['sexo clasif'] = ''
violencia.loc[violencia['sexo_'] == 'Masculino','sexo clasif'] = 1
violencia.loc[violencia['sexo_'] == 'Femenino','sexo clasif'] = 2

violencia['seg social clasif'] = ''
violencia.loc[violencia['Tipo de Seguridad Social'] == 'Contributivo','seg social clasif'] = 1
violencia.loc[violencia['Tipo de Seguridad Social'] == 'Especial','seg social clasif'] = 2
violencia.loc[violencia['Tipo de Seguridad Social'] == 'Excepción','seg social clasif'] = 3
violencia.loc[violencia['Tipo de Seguridad Social'] == 'Indeterminado','seg social clasif'] = 4
violencia.loc[violencia['Tipo de Seguridad Social'] == 'No Afiliado','seg social clasif'] = 5
violencia.loc[violencia['Tipo de Seguridad Social'] == 'No asegurado','seg social clasif'] = 6
violencia.loc[violencia['Tipo de Seguridad Social'] == 'Subsidiado','seg social clasif'] = 7

violencia['sexo agresor clasif'] = ''
violencia.loc[violencia['sexo_agre'] == 'F','sexo agresor clasif'] = 1
violencia.loc[violencia['sexo_agre'] == 'I','sexo agresor clasif'] = 2
violencia.loc[violencia['sexo_agre'] == 'M','sexo agresor clasif'] = 3
violencia.loc[violencia['sexo_agre'] == 'SD','sexo agresor clasif'] = 4
```

```python
violencia.head(2)
```

```python
violencia.describe()
```

```python
violencia.describe().round(1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(0)
plt.hist(violencia['naturaleza'],edgecolor = 'black', linewidth=1)
plt.title('Naturaleza')

plt.figure(1)
plt.hist(violencia['edad_agre'],edgecolor = 'black', linewidth=1)
plt.title('Edad agresor')

plt.figure(2)
plt.hist(violencia['actividad'],edgecolor = 'black', linewidth=1)
plt.title('Actividad')
```

```python
violencia_corr = violencia.corr(method='spearman')
violencia_corr
```

```python
sns.heatmap(violencia_corr,xticklabels=violencia_corr.columns,yticklabels=violencia_corr.columns,cmap='coolwarm')
```

```python
plt.figure(0)
sns.boxplot(x='naturaleza',y='año', data=violencia)
plt.show()

plt.figure(1)
sns.boxplot(x='sexo clasif',y='año', data=violencia)
plt.show()

plt.figure(2)
sns.boxplot(x='Grup edad clasif',y='año', data=violencia)
plt.show()

plt.figure(3)
sns.boxplot(x='Ciclo de vida clasif',y='año', data=violencia)
plt.show()

plt.figure(4)
sns.boxplot(x='seg social clasif',y='año', data=violencia)
plt.show()

plt.figure(5)
sns.boxplot(x='sexo agresor clasif',y='año', data=violencia)
plt.show()

#Grup edad clasif	Ciclo de vida clasif	sexo clasif	seg social clasif	sexo agresor clasif
```
